```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("avila-ts.txt",sep = ",", header=None)
```

```python
df.columns =['intercolumnar_distance', 'upper_margin', 'lower_margin', 'exploitation','row_number','modular_ratio',  'interlinear_spacing','weight','peak_number','diff','class']
```

```python
df.describe()
```

```python
list_var = df.columns[:-1]
fig = plt.figure()

for i in list_var:
    fig, ax = plt.subplots(figsize=(10,8))
    ax = sns.boxplot(x='class',y=i,data=df,whis=150)
plt.show()
```

```python

attributes_correlation = df.corr().abs()

plt.figure(figsize=(12, 10))
sns.heatmap(attributes_correlation, vmax=1, vmin=0, annot=True, square=True, linewidths=.5, cmap='YlOrBr')
plt.title('Correlation des attributs')
plt.show()
```

```python
df.class.plot.pie(autopct="%.1f%%");
```
